```python
# Create an API client
from anthropic import Anthropic


client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
# Make a request
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "What is quantum computing? Answer in 1 sentence."
        }
    ]
)  
```

```python
# Helper functions to manage messages
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)
    
def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)
    
def chat(messages):
    response = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages
    )
    return response.content[0].text
```

```python
# Make a starting list of messages
messages = []

# Add in the initial user message
add_user_message(messages, "What is quantum computing? Answer in 1 sentence.")

# Pass the list of messages into the chat function to get a response
answer = chat(messages)

# Take the answer and add it to the list of messages
add_assistant_message(messages, answer)

# Add in the user's follow-up question
add_user_message(messages, "Can you explain that more with another sentence?")

# Call chat again to get a response to the follow-up question
answer = chat(messages)
answer
```

```python
# Make an initial list of messages
messages = []

# Use a "while true" loop to keep the conversation going until the user types "exit"
while True:
    # Get user input
    user_input = input("> ")
    print(">", user_input)
    
    # Add the user's message to the list of messages
    add_user_message(messages, user_input)
    # Call Claude with the "chat
    answer = chat(messages)
    # Add generated text to the list of messages
    add_assistant_message(messages, answer)
    # Print the generated text
    print("---")
    print(answer)
    print("---")
   
```
